# Seaborn factorplot

Factor plot is useful for displaying multiple plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

sns.set()
```

```python
titanic = sns.load_dataset("titanic")
```

```python
sns.factorplot(data=titanic, col="survived", row="class", y="age", x="embarked", kind="violin", hue="sex")
```
