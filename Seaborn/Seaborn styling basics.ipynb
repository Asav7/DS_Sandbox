```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
# Load iris data
iris = sns.load_dataset("iris")
```

```python
# Construct iris plot
sns.swarmplot(x="species", y="petal_length", data=iris)

# Show plot
plt.show()
```

```python
# Too make it look better
sns.set()
sns.swarmplot(x="species", y="petal_length", data=iris)
plt.show()
```

```python
sns.set(rc={"font.size":6,"axes.labelsize":8})
sns.swarmplot(x="species", y="petal_length", data=iris)
plt.show()
```

You can use **axes_style()** in a **with** statement to temporarily set the plot style. This, in addition to the use of plt.subplot(), will allow you to make figures that have differently-styled axes, like in the example below:

```python
# Load data
iris = sns.load_dataset("iris")
tips = sns.load_dataset("tips")

# Set axes style to white for first subplot
with sns.axes_style("white"):
    plt.subplot(211)
    sns.swarmplot(x="species", y="petal_length", data=iris)

# Initialize second subplot
plt.subplot(212)

# Plot violinplot
sns.violinplot(x = "total_bill", data=tips)

# Show the plot                   
plt.show()
```

To set the figure size (for axes level functions), you can make use of the plt.subplots() function to which you pass the figsize argument.

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.swarmplot(x="species", y="petal_length", data=iris, ax=ax)
plt.show()
```

To set the figure size (for axes level functions), you rely on two parameters to set the Figure size: **size** and **aspect**

```python
titanic = sns.load_dataset("titanic")
g = sns.factorplot("class", "survived", "sex", data=titanic, kind="bar", size=4, aspect=2, palette="muted", legend=False)
plt.show()
```
