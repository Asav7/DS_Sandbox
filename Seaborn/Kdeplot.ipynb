# Seaborn Kdeplot

```python
#Imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline

#styling
sns.set()
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.kdeplot(tips.total_bill, shade=True)
```

**Two dimensional kernel density estimation**

```python
sns.kdeplot(tips.total_bill, tips.tip)
```

```python
sns.kdeplot(tips.total_bill, tips.tip, shade=True)
```

```python
sns.kdeplot(tips.total_bill, tips.tip, n_levels=25)
```

```python
sns.kdeplot(tips.total_bill, shade=True, cumulative=True) # cdf instead of pdf
```
