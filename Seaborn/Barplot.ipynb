# Seaborn Barplot

```python
#Imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline

#styling
sns.set()
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.barplot(data=tips, x="day", y="tip")
```

**with hue attribute**

```python
sns.barplot(data=tips, x="day", y="tip", hue="smoker")
```

To create **horizontal bar plot** replace **x** for **y**

```python
sns.barplot(data=tips, x="tip", y="day")
```

The default aggregate function is mean, to select other function use **estimator** argument

```python
sns.barplot(data=tips, x="day", y="tip", hue="smoker", estimator=np.sum, palette="summer")
```

### Confindence intervals for error bars
To specify confidence intervals for error bars use **ci** attribute

```python
sns.barplot(data=tips, x="day", y="tip", ci=90)
```

```python
sns.barplot(data=tips, x="day", y="tip", ci=10)
```

Capsize for errorbars

```python
sns.barplot(data=tips, x="day", y="tip", ci=80, capsize=0.2)
```
