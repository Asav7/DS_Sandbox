# Seaborn Boxplot

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.set()
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.boxplot(data=tips, x="total_bill")
```

By default whiskers are 1,5 interquartile ranges long. They are shorter if there are non datapoints behind them.

```python
data = pd.Series([10, 10, 10, 11, 13, 9, 8, 20])
sns.boxplot(x=data).set(xlim=[7, 21])
```

```python
sns.boxplot(data=tips, x="total_bill", notch=True) # notch adds confidence intervals for the median
```

Confidence interwal is estimated with bootstrapping.

```python
sns.boxplot(data=tips, y="total_bill", x="day") 
```

```python
sns.boxplot(data=tips, y="total_bill", x="day", hue="smoker") 
```

```python
fig, ax = plt.subplots(figsize=(6,5))
sns.boxplot(data=tips, y="total_bill", x="day") 
sns.swarmplot(data=tips, y="total_bill", x="day", color=".3") 
```
