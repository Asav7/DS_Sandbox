# Seaborn jointplot

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.set()
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip")
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="reg")
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="hex")
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="kde")
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip").plot_joint(sns.kdeplot, n_levels=6)
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip").plot_joint(sns.kdeplot, n_levels=6).plot_marginals(sns.rugplot)
```
