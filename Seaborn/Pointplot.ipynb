# Seaborn pointplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

sns.set()
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.pointplot(data=tips, x="time", y="total_bill")
```

Pointplot above shows mean at lunch and mean at dinner with 90% confidence interwal between them. The line that connects means helps to judge relative heights.

```python
sns.pointplot(data=tips, x="time", y="total_bill", hue="smoker", dodge=True)
```

While plotting several graphs using *hue* attribute it's recommended to use *dodge=True* for better readability

```python
sns.pointplot(data=tips, x="time", y="total_bill", hue="smoker", markers=["o", "x"], linestyles=["-", "--"])
```

```python
sns.pointplot(data=tips, x="time", y="total_bill", hue="smoker", estimator=np.median)
```

```python
sns.pointplot(data=tips, x="time", y="total_bill", hue="smoker", ci=50)
```

```python
sns.pointplot(data=tips, x="time", y="total_bill", hue="smoker", capsize=.2)
```
