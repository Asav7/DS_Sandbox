```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
sns.set()
%matplotlib inline
```

```python
x = np.linspace(0, 15, 50)
data = np.sin(x) + np.random.rand(10, 50) + np.random.randn(10, 1)
sns.tsplot(data=data)
```

```python
gammas = sns.load_dataset("gammas")
gammas.head(3)
```

```python
sns.tsplot(data=gammas, time="timepoint", value="BOLD signal", unit="subject", condition="ROI")
```
