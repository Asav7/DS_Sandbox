# Seaborn violinplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

sns.set()
```

```python
tips = sns.load_dataset("tips")
```

Violinplot combines box plot with the KDE.

```python
sns.violinplot(data=tips, x="total_bill") # by default inner="box"
```

```python
sns.violinplot(data=tips, x="total_bill", inner="quartile")
```

```python
sns.violinplot(data=tips, x="total_bill", inner="sticks")
```

```python
sns.violinplot(data=tips, x="total_bill", inner="points")
```

```python
sns.violinplot(data=tips, y="total_bill", x="day") #by default scale="area"
```

```python
sns.violinplot(data=tips, y="total_bill", x="day", scale="width")
```

```python
sns.violinplot(data=tips, y="total_bill", x="day", scale="count") #allows to visualize how data is distributed in the dataset
```

```python
sns.violinplot(data=tips, y="total_bill", x="day", scale="count", hue="smoker")
```

```python
sns.violinplot(data=tips, y="total_bill", x="day", scale="count", hue="smoker", split=True)
```
