# Seabor stripplot

```python
#Imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline

#styling
sns.set()
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.stripplot(data=tips, x="total_bill", y="day")
```

```python
sns.stripplot(data=tips, x="total_bill", y="day", jitter=True) # use jitter to spread data points
```

```python
sns.stripplot(data=tips, y="total_bill", x="day", jitter=True, hue="smoker", dodge=True) # while using hue you can also \ 
                                                                                          # use dodge to group datapoints
```

```python
# you can use stripplot with boxplot
sns.boxplot(data=tips, x="total_bill", y="day")
sns.stripplot(data=tips, x="total_bill", y="day", color="0.3", jitter=0.05, )
```

### Seaborn swarmplot

Swarmplot is similar to stripplot, but spreads datapoints so they won't overlap.

```python
sns.swarmplot(data=tips, x="total_bill", y="day")
```
