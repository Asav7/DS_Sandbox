# Seaborn lmplot 

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
sns.set()
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip")
```

The confidence interval is dony by bootstrap. Which means we sample with replacement from our datapoints, then we compute linear regression. We do this a number of times, and then we use the farthest ones to draw the confidence interval.

```python
sns.lmplot(data=tips, x="total_bill", y="tip", hue="smoker")
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="smoker")
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="smoker", row="sex")
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="smoker", row="sex", hue="time")
```

```python
sns.lmplot(data=tips, x="size", y="tip")
```

```python
sns.lmplot(data=tips, x="size", y="tip", x_jitter=.05) # add jitter to improve readability
```

```python
sns.lmplot(data=tips, x="size", y="tip", x_estimator=np.mean) # confidence interwals done by bootstrap
```

```python
sns.lmplot(data=tips, x="size", y="tip", x_estimator=np.mean, ci=50)
```

```python
sns.lmplot(data=tips, x="size", y="tip", x_estimator=np.mean, ci=50, x_ci=98) # x_ci controls confidence interval for x estimator
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", x_bins=10) # its possible to group continuous data into bins
```
