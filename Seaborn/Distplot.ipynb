# Seaborn Dist Plot

```python
#Imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline

#styling
sns.set()
```

```python
tips = sns.load_dataset("tips")
sns.distplot(tips.total_bill)
```

```python
sns.distplot(tips.total_bill, rug=True) # draws a line on the bottom for every datapoint
```

```python
sns.distplot(tips.total_bill, rug=True, norm_hist=False, kde=False) #turn of kde and set norm_hist=False to get not normed histogram
```

To fit a specific distribution to our data use **fit** argument

```python
import scipy.stats as stats
sns.distplot(tips.total_bill, rug=True, fit=stats.gamma)
```
