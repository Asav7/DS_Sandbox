# Seaborn Pairplot

```python
#Imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline

#styling
sns.set()
```

```python
iris = sns.load_dataset("iris")
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue="species", markers=["o", "s", "D"])
```

```python
sns.pairplot(iris, kind="reg")
```

```python
sns.pairplot(iris, kind="reg", hue="species", diag_kind="kde")
```
