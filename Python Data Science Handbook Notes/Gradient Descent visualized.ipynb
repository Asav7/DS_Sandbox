```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
plt.style.use("ggplot")
```

```python
x_data = np.linspace(0, 7, 100)
y_data = 2 + x_data + np.random.normal(size=100)
```

```python
plt.scatter(x_data, y_data)
```

hipothesis function

\begin{equation*}
h(x) = \theta_0 + \theta_1  x
\end{equation*}

```python
def cost(theta0, theta1):
    return 1 / (2 * len(x_data)) * sum((theta0 + theta1 * x_data - y_data) ** 2)
```

```python
def hypothesis(theta0, theta1):
    return theta0 + theta1 * x_data
```

```python
plt.scatter(x_data, y_data)
plt.plot(x_data, hypothesis(2, 1), c='b')
```

```python
thetas0 = np.linspace(0, 3, 20)

thetas1 = np.linspace(0, 2, 20)
```

```python
th0, th1 = np.meshgrid(thetas0, thetas1)
```

```python
costs = np.array(np.sum([((th0 + th1 * x) - y) ** 2 for x, y in zip(x_data, y_data)], axis=0))

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(th1, th0, costs)
plt.show()
```
