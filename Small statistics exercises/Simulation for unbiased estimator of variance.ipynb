# Simulation for unbiased estimator of variance

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

plt.style.use("ggplot")
```

```python
data = np.random.uniform(0, 100, 5000)
```

```python
plt.hist(data)
```

```python
def var_n(data, param=0):
    return np.sum((np.square(np.subtract(data, data.mean()))))/(len(data) + param)
```

```python
def plot_estimates(data, param=0, n=30):
    var_est = np.array([])
    for i in range(n):
        sample = np.random.choice(data, 40)
        var_est = np.append(var_est, var_n(sample, param=param))
        plt.scatter(i, np.mean(var_est))
    plt.plot([i for i in range(n)], [data.var() for i in range(n)])
```

```python
plot_estimates(data, n=500)
```

```python
plot_estimates(data, param=-1, n=500)
```

```python
plot_estimates(data, param=(-2), n=500)
```
