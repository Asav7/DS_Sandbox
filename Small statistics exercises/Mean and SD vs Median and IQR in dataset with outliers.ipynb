# Mean and SD vs median and IQR in dataset with outliers

```python
import numpy as np
import scipy.stats as stats
from matplotlib import pyplot as plt

%matplotlib inline
plt.style.use("ggplot")
```

```python
data = [35, 50, 50, 50, 56, 60, 60, 75, 250]
data = np.array(data)
```

```python
mean = data.mean()
std = data.std()
median = np.median(data)
iqr = stats.iqr(data, interpolation='midpoint')
print("Mean: ", mean)
print("SD: ", std)
print("Median: ", median)
print("IQR: ", iqr)
```

```python
plt.hist((data - mean)/std, normed=True)
x = np.linspace(-5, 5, 300)
plt.plot(x, stats.norm.pdf(x))
```

```python
plt.hist((data - median)/iqr, normed=True)
plt.plot(x, stats.norm.pdf(x))
```

```python
import seaborn as sns
sns.kdeplot(data)
plt.hist(data, normed=True)
```
