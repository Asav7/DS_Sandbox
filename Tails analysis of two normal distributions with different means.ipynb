```python
import scipy as sp
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use("ggplot")
%matplotlib inline
```

```python
pdf = sp.stats.norm.pdf
```

```python
d1 = 0.737374
d2 = 1.141414
delta = d2 - d1
```

```python
x = np.linspace(-4, 5, 100)
x_1 = np.linspace(-3 - delta, 3 - delta)
plt.plot(x, pdf(x))
plt.plot(x, pdf(x, loc=-0.4))
plt.gca().set_xlim(-3, 3)
```

```python
plt.plot(x, pdf(x))
plt.plot(x, pdf(x, loc=-0.4))
ax = plt.gca()
ax.set_xlim(2.4, 3.5)
ax.set_ylim(top=0.03)
```

```python
start = 3
stop = 2.6
delta_m = pdf(start) - pdf(stop)
delta_k = pdf(start, loc=-0.4) - pdf(stop, loc=-0.4)
print("M: " + str(delta_m/pdf(start)))
print("K: " + str(delta_k/pdf(start)))
```
