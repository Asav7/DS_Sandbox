## Tails of normal distributions with different means

```python
import scipy as sp
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use("ggplot")
%matplotlib inline
```

```python
pdf = sp.stats.norm.pdf
x = np.linspace(-5, 5, 1000)
distribution = pdf(x)
dist_lower_mean = pdf(x, loc=-0.4)
```

```python
plt.plot(x, distribution)
plt.plot(x, dist_lower_mean)
plt.gca().set_xlim(-3, 3)
```

**Zoming on the right tail**

```python
plt.plot(x, distribution)
plt.plot(x, dist_lower_mean)
ax = plt.gca()
ax.set_xlim(2.4, 3.5)
ax.set_ylim(top=0.03)
```

```python
plt.plot(x, distribution/dist_lower_mean)
plt.gca().set_xlim(-5, 5)
```

---

```python
plt.semilogy(x, distribution)
plt.semilogy(x, dist_lower_mean)
plt.gca().set_xlim(1, 5)
```

```python
x = np.linspace(-20, -5, 100)
plt.plot(x, x+2)
plt.plot(x, x+10)
plt.plot(x, (x+2)/(x+10))
plt.plot(x, x*0, c='k')
ax = plt.gca()
ax.set_xlim(-20, -5)
ax.set_ylim(-20, 20)
```

```python
size = 30000
bins = 80
n1 = np.random.normal(size=size)
n2 = np.random.normal(loc=-1, size=size)
plt.hist(n1[n1>1], bins=bins)
plt.hist(n2[n2>1], bins=bins)
plt.gca().set_xlim(left=1)
plt.gca().set_ylim()
```
