# Logistic distribution
https://en.wikipedia.org/wiki/Logistic_distribution

In probability theory and statistics, the **logistic distribution** is a continuous probability distribution. Its cumulative distribution function is the logistic function, which appears in logistic regression and feedforward neural networks. It resembles the normal distribution in shape but has heavier tails (higher kurtosis).

The probability density function (pdf) of the **logistic distribution** is given by:

$
\begin{align}
f(x; \mu, s) = \frac{e^{-\frac{x - \mu}{s}}}{s\left( 1 + e^{-\frac{x - \mu}{s}}\right)^2}
\end{align}
$

```python
from matplotlib import pyplot as plt
import numpy as np
import scipy.stats as stats

%matplotlib inline
plt.style.use('ggplot')
```

```python
x = np.linspace(-4, 4)
plt.plot(x, stats.norm.pdf(x))
plt.plot(x, stats.logistic.pdf(x))
plt.legend(['normal', 'logistic'])
```

# Logistic function
https://en.wikipedia.org/wiki/Logistic_function

A **logistic function** or **logistic curve** is a common "S" shape (sigmoid curve), with equation:

$
\begin{align}
f(x) = \frac{L}{1+e^{-k(x-x_{0})}}
\end{align}
$

$e$ = the natural logarithm base (also known as Euler's number), <br>
$x_0$ = the x-value of the sigmoid's midpoint,<br>
$L$ = the curve's maximum value, and <br>
$k$ = the steepness of the curve

```python
def f(x, L=1, e=2.718281828459045, k=1, x0=0):
    return L/(1+e**(-k*(x-x0)))
```

```python
x = np.linspace(-5, 5, 100)
```

```python
plt.plot(x, f(x))
```

## $x_0$ change

```python
plt.plot(x, f(x), linewidth=2)
plt.plot(x, f(x, x0=-1))
plt.plot(x, f(x, x0=2))
plt.legend(['x0 = 0', 'x0 = -1', 'x0 = 2'], loc=2)
```

## $L$ change

```python
plt.plot(x, f(x), linewidth=2)
plt.plot(x, f(x, L=0.5))
plt.plot(x, f(x, L=3))
plt.legend(['L = 1', 'L = 0.5', 'L = 3'], loc=2)
```

## $k$ change

```python
plt.plot(x, f(x), linewidth=2)
plt.plot(x, f(x, k=0.5))
plt.plot(x, f(x, k=4))
plt.legend(['k = 1', 'k = 0.5', 'k = 4'], loc=2)
```

### logistic function in logistic regression

The logistic function is useful because it can take any real input $t$, ($t\in \Bbb{R}$), whereas the output always takes values between zero and one and hence is interpretable as a probability. The logistic function $\sigma (t)$ is defined as follows:

$
\begin{align}
\sigma(t) = \frac{e^{\ t}}{e^{\ t} - 1} =\frac{1}{1 + e^{-t}}
\end{align}
$

Let us assume that $t$ is a linear function of a single explanatory variable $x$ (the case where $t$ is a linear combination of multiple explanatory variables is treated similarly). We can then express $t$ as follows:

$
\begin{align}
t = \beta_0 + \beta_1 x
\end{align}
$

And the logistic function can now be written as:

$
\begin{align}
F(x) = \frac{1}{1 + e^{\ \beta_0 \ + \ \beta_1 x}}
\end{align}
$
